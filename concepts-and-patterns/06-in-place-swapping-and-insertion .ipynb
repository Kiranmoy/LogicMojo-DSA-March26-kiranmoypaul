# In-place by Swapping + Insertion (Boundary method)

**Problem**: Given two sorted arrays, `X[]` and `Y[]` of size m and n each, merge elements of `X[]` with elements of array `Y[]` by maintaining the sorted order, i.e., fill `X[]` with the first m smallest elements and fill `Y[]` with remaining elements.

**Pattern:**
* Iterate X from the end and Y from the beginning.
* If `X[i] > Y[0]`, swap them, then insert the swapped `Y[0]` into its correct spot in Y (since Y must remain sorted).

**Intuition:**
* The largest elements of X that actually belong to Y should be pushed to Y.
* The smallest element of Y that belongs to X takes the vacated spot.
* After each swap, you maintain Y sorted via insertion (shift).

**Why it works:**
* You keep invariant: X is non-decreasing, and Y is non-decreasing, and no element in X should be greater than the smallest element in Y.
* Repeating this moves out-of-place large elements from X into Y and pulls in the smallest required elements from Y.

**Complexity:**
* Worst-case Time: `O(m * n)` due to insertion into Y (shifting).
* Space: `O(1)`

# Solution

```python
# Function to in-place merge two sorted arrays X[] and Y[]
def merge_in_place_boundary(X, Y):
    m, n = len(X), len(Y)

    # Consider each element X[i] of array X with first element of Y
    # ignore the element, if it is already in the correct order;
    # otherwise, swap it with the smaller element, which happens to be the first element of Y.
    for i in range(m):
        # compare the current element of X[] with the first element of Y[]
        if X[i] > Y[0]:
            # Swap X[i] with Y[0]
            X[i], Y[0] = Y[0], X[i]

            # Restore sorted order of Y by inserting Y[0] at proper place
            first = Y[0]
            k = 1
            # Shift all smaller elements to the left, until an element > first is found
            while k < n and Y[k] < first:
                Y[k - 1] = Y[k]
                k += 1
            # insert first at its proper place
            Y[k - 1] = first

X = [ 1, 4, 7, 8, 10 ]
Y = [ 2, 3, 9 ]
merge_in_place_boundary(X, Y)

print(X)
print(Y)
```
