# Problem - Partitioning Negative Before Positive

**Problem**: An array contains both positive and negative numbers in random order. Rearrange the array elements so that all negative numbers appear before all positive numbers.
Examples : 

**Input**: `[-12, 11, -13, -5, 6, -7, 5, -3, -6]`

**Output**: `[-12, -13, -5, -7, -3, -6, 11, 6, 5]`


# Core Pattern - Partitioning

* Choose "negative" as the pivot category.
* Move all negative numbers to the left region.
* Move all positive numbers to the right region.
* Done in `O(n)` time and `O(1)` extra space.

# Intuition

**Think of the array as two zones:**

```
[ NEGATIVE ZONE | POSITIVE ZONE ]
```

* You scan the array from left to right.
* Whenever you see a **negative number**, you simply **expand the negative zone** by swapping it into the next free left slot.
* This is the same mental model used when partitioning numbers around a pivot.

# Solution

````python
def rearrange_negative_postive(arr):
    """
    Rearranges arr in-place so that all negative numbers are on the left
    and all non-negative (positive/zero) numbers are on the right.
    """

    # Initialize two pointers, left & right
    left = 0                # points to the start of the array
    right = len(arr) -1     # points to the end of the array

    
    while left <= right:
        # Both left and right are negative -> left is already in the correct region
        if arr[left] < 0 and arr[right] < 0:
            left += 1

        # Left is non-negative, right is negative -> swap to correct sides
        elif arr[left] >= 0 and arr[right] < 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1

        # Both are non-negative -> shrink right side
        elif arr[left] >= 0 and arr[right] >= 0:
            right -= 1

        # Left is negative, right is non-negative -> both on the correct side
        else:
            left += 1
            right -= 1


arr = [-12, 11, -13, -5, 6, -7, 5, -3, -6]
rearrange_negative_postive(arr)
print(arr)  # e.g., [-12, -6, -13, -5, -3, -7, 5, 6, 11] (order within groups may vary)
````
