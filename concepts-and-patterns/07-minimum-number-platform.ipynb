# Problem - Minimum Number Platform

Given a schedule containing the arrival and departure time of trains in a station, find the minimum number of platforms needed to avoid delay in any train’s arrival.

**Example:**

```
Trains arrival = [2.00, 2.10, 3.00, 3.20, 3.50, 5.00]
Trains departure = [2.30, 3.40, 3.20, 4.30, 4.00, 5.20]
```
The minimum platforms needed are 2.


# Core Pattern

**Sort both arrival and departure times independently.**

Then use **two pointers**, one for arrival and one for departure:
* If a train **arrives before** the earliest departure → we **need another platform**.
* If a train **arrives after or exactly** at the earliest departure → we can **free a platform**.

This is essentially a **sweep-line algorithm**, used in many **interval overlap** problems.

# Intuition Behind the Pattern

**Think of time flowing from left → right.**

You stand at the platform and track:
* **Arrivals** cause an increment (more trains → more platforms needed)
* **Departures** cause a decrement (a platform becomes free)

This is exactly like tracking how many “intervals” overlap at any point in time.

# Solution

````python
def find_max_platform(arrivals, departures):
    # Sort arrivals & departures times
    arrivals.sort()
    departures.sort()

    # two pointers to track for arrival and departure time
    a = 0 # arrival pointer
    d = 0 # departure pointer

    # track platform
    p = 0

    # max platform needed
    max_platform = 0

    # run till all trains have arrived
    while a < len(arrivals):

        # if a train is scheduled to arrive next
        if arrivals[a] < departures[d]:
            p = p + 1   # increase the count of platforms and update the minimum platforms if required
            a = a + 1   # move the pointer to the next arrival
            max_platform = max(max_platform,p)

        # if the train is scheduled to depart next
        # Or, if two trains are arriving and departing simultaneously
        else:
            p = p - 1   # decrease platform count as train will depart
            d = d + 1   # move pointer to the next departure time

    return max_platform

arrivals = [2.00, 2.10, 3.00, 3.20, 3.50, 5.00]
departures = [2.30, 3.40, 3.20, 4.30, 4.00, 5.20]

print(find_max_platform(arrivals,departures))
````

# Complexity

* Sorting: `O(n log n)`
* Two-pointer scan: `O(n)`
* Total Time Complexity: `O(n log n)`
* Total Space Complexity: `O(1)` (in-place)
