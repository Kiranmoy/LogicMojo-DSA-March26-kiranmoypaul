# Problem - Maximum Product Subarray

Given an array that contains both positive and negative integers, find the product of the maximum product subarray.

**Input**: `arr = [6, -3, -10, 0, 2]`

**Output**:   `180`  The subarray is `[6, -3, -10]`


# Pattern - Dynamic Sliding Window with “Track Max & Min”

This problem uses a unique DP-style pattern:
* Because multiplying by a negative number flips signs.
* You must track both the maximum and the minimum product at every step.

**1. A negative number can flip signs**
* Multiplying by a negative turns big positive → big negative
* But multiplying by another negative can turn it → big positive again

So a negative can help or hurt.

**2. Zero breaks the product**
* Any subarray containing 0 becomes 0.
* After zero, we must “restart” the product.

**3. We need to track BOTH max and min (because of negatives)**

So at every index:
* **max_ending_here** → largest product ending at this index
* **min_ending_here** → smallest product (most negative) ending at this index
* When we see a negative number: → **we swap them**


> **Key Insight**: Because of negatives, the smallest (most negative) product can suddenly become the largest when multiplied by another negative.

# Solution

```python
def maxProductSubarray(nums):
    
    max_ending = nums[0]     # max positive product ending at the current position
    min_ending = nums[0]     # min positive product ending at the current position
    max_product = nums[0]    # Initialize overall max product

    # Traverse through the array
    for num in nums[1:]:

        # Multiplying by a negative makes an existing negative product positive & existing positive product negative
        # So existing max product & min product gets swapped
        # max product -> becomes minimum
        # min product -> becomes maximum
        if num < 0:
            max_ending, min_ending = min_ending, max_ending

        # Handling 0 - restart the product after encountering num=0
        max_ending = max(num, max_ending * num)
        min_ending = min(num, min_ending * num)

        max_product = max(max_product, max_ending)

    return max_product


# Example
num = [6, -3, -10, 0, 2]
print(maxProductSubarray(num))  # 180
```
