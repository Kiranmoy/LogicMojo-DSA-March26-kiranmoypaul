# Problem

You are given a read-only array of **N** integers with values also in the range `[1,N]`, both inclusive. Each integer appears exactly once except **A**, which appears twice, and **B**, which is missing. The task is to find the repeating and missing numbers **A** and **B** where **A** repeats twice and **B** is missing.

**Input Format**
* First line is the length of the array - n
* The second line contains n integers that are elements of the array

**Constraints**

```
1 <= n <= 105
1 <= arr[i] <= n
```

**Output Format**: 

Print an array of integers containing repeating and missing numbers, respectively

**Sample Input**

```
3
3 1 3
```

**Sample Output**

```
3 2
```

**Explanation**

A = 3, B = 2. Since 3 is appearing twice and 2 is missing

# Math method (Sum & Sum of Squares)

**Intuition**

* If nothing was wrong, the sum would be `S = 1 + 2 + ... + N`, and the sum of squares would be `P = 1² + 2² + ... + N²`.
* But the array has A twice and B missing, so:
    * The sum becomes `S_arr = S + A - B  →  Δ1 = S_arr - S = A - B`
    * The sum of squares becomes `P_arr = P + A² - B²  →  Δ2 = P_arr - P = A² - B² = (A - B)(A + B)`
* **From these:**
    * A - B = Δ1
    * A + B = Δ2 / Δ1
* Solve the two equations to get **A** and **B**.

**Tiny example**

```
arr = [3, 1, 2, 5, 3] (N=5) → A=3 (repeats), B=4 (missing)
```

# Solution

````python
def find_repeating_missing_math(arr):
    n = len(arr)
    # Expected sums if the array had 1..N with no issues
    S = n * (n + 1) // 2                      # 1 + 2 + ... + N
    P = n * (n + 1) * (2*n + 1) // 6          # 1^2 + 2^2 + ... + N^2

    # Actual sums from the given array
    S_arr = sum(arr)
    P_arr = sum(x * x for x in arr)

    # Differences caused by A repeating and B missing
    delta1 = S_arr - S                         # = A - B
    delta2 = P_arr - P                         # = A^2 - B^2 = (A-B)(A+B)

    # Avoid division by zero (shouldn't happen under problem guarantees)
    if delta1 == 0:
        return None

    # A + B
    sumAB = delta2 // delta1

    # Solve:
    # A = ( (A+B) + (A-B) ) / 2
    # B = ( (A+B) - (A-B) ) / 2
    A = (sumAB + delta1) // 2
    B = sumAB - A

    # Debug prints to understand step-by-step
    print(f"N={n}")
    print(f"Expected S={S}, P={P}")
    print(f"Actual   S_arr={S_arr}, P_arr={P_arr}")
    print(f"delta1=A-B={delta1}, delta2=A^2-B^2={delta2}")
    print(f"A+B={sumAB}")
    print(f"Result -> A(repeating)={A}, B(missing)={B}")

    return A, B

# Try the example:
find_repeating_missing_math([3, 1, 2, 5, 3])
````
